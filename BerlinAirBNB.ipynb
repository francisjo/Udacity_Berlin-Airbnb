```python
def features_categorical(df):
    print("List of Categorical Columns :\n")
    percent_uniquevaluesper_column = df[list(df.select_dtypes(include=['object']))].nunique()/len(df) * 100
    percent_uniquevaluesper_column.sort_values(ascending=False).plot(kind="bar",figsize=(20,5))
    plt.title("Percentage of Unique values per each column")
    plt.xlabel('Column Name')
    plt.ylabel('Percentage of Unique values')
    plt.show()
    for x in list(df.select_dtypes(include=['object'])):
        if (df[x].nunique() < 10):
            uniquelist = df[x].unique()
            print("- ",x,":", df[x].nunique()," and list of unique values that are less than 10 : ", uniquelist)
```

```python
def assesing_df(df):
    name =[x for x in globals() if globals()[x] is df][0]
    # Print rows and columns in dataset
    print( "Size of", name, ": \n"
          "number of rows: ",df.shape[0],"\n"
          "number of columns: ",df.shape[1])
    # Assess null values
    percent_missing_per_column =(df.isnull().sum()/len(df)*100)
    percent_missing_per_column.hist(bins=10,xlabelsize=8)
    plt.grid(True)
    plt.title("Distribution of percentage of missing values in each column")
    plt.ylabel('Number of Columns')
    plt.xlabel('Percentage of Missing Values in Columns')
    plt.show()
  
    # Describ
    #print(df.describe(include='all'))
    percent_missing_per_column.plot(kind="bar", figsize=(20,5))
    plt.title("Percentage of missing values per each column")
    plt.xlabel('Column Name')
    plt.ylabel('Percentage of Missing Values')
    outliers  = percent_missing_per_column[percent_missing_per_column>20]
    plt.grid(True)
    plt.show()
    if len(outliers) == 0 :
        print('\n There are not columns that have more than 20% Missing Values')
    else:
        outliers.sort_values(ascending=False).plot(kind="bar")
        plt.title("List of columns have more than 20% Missing Values")
        plt.xlabel("Column name")
        plt.ylabel("Number of missing values")
        plt.grid(True)
        plt.show()
    percent_missing_val_row = df.isnull().sum(axis=1)/(len(df.columns)) * 100
    plt.hist(percent_missing_val_row, bins=10, facecolor='b', alpha=0.75)
    plt.xlabel('Percentage of Missing Values')
    plt.grid(True)
    plt.ylabel('Number of Rows')
    plt.title('Percentage of Missing Values in Rows')
    plt.show()
    features_categorical(df)
```
